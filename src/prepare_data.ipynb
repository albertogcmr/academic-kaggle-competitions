```python
from clean_data import clean_data
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('../cars-competition/data/cars_train.csv', index_col='Id')
data.head()
```

```python
train = clean_data(data)
train.head()
```

```python
# Select only relevant columns
train = train[['year','manufacturer','condition','cylinders','fuel','odometer','title_status','transmission','drive','size','price']]
```

```python
train.dtypes
```

```python
# Getting dummies for categorical columns
train = pd.get_dummies(train,drop_first=True)
```

```python
# Get X and y
X = train.drop(columns='price')
y = train['price']
```

```python
# Feature reduction
pca = PCA(n_components = 30)
X_new = pca.fit_transform(X)
```

```python
pd.DataFrame(X_new)
```

```python
# Split train and test
X_train, X_test, y_train, y_test = train_test_split(X_new,y,test_size=0.2)
```

```python
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
# Predict
y_pred = lin_reg.predict(X_test)
```

```python
# Check error
mean_squared_error(y_test, y_pred)
```

```python
# Check error
mean_squared_error(y_test, [y_test.mean()]*len(y_test))
```

```python
# Random Forrest
regr = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
regr.fit(X_train,y_train)
```
