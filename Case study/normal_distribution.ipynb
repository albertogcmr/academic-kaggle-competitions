```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from numpy.random import normal
```

```python
# Read files
train = pd.read_csv('../cars-competition/data/cars_train.csv', index_col='Id')
submission = pd.read_csv('../cars-competition/data/cars_test.csv', index_col='Id')
```

```python
X = train['price']
avg = X.mean()
stdv = X.std()/10
```

```python
avg, stdv
```

```python
y_sub30 = normal(loc=avg, scale=stdv, size=submission.shape[0])
y_sub30 = abs(y_sub30).astype('int')
```

```python
list(y_sub30)
```

```python
# Prepare for submission
sub30 = submission
sub30['price'] = y_sub30
sub30 = sub30['price'] 
sub30.to_csv('../cars-competition/data/normaldist.csv', header=True, index=True)
```
